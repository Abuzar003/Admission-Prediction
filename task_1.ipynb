<h1>Data Preprocessing</h1>


```python
 df.isnull().values.sum()
```

```python
from sklearn.preprocessing import StandardScaler
x = df.drop(columns=['Serial No.', 'Chance of Admit '])
y = df['Chance of Admit ']
scaler_standard = StandardScaler()
standered_df = scaler_standard.fit_transform(x)
df_standard_scaled = pd.DataFrame(standered_df, columns=x.columns)
df_standard_scaled.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df_standard_scaled, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```
